# Load data

```python
bay_area_counties = [
    'Alameda', 'Contra Costa', 'Marin', 'San Francisco', 'San Mateo', 'Santa Clara', 'Napa', 'Sonoma', 'Solano'
]
```

```python
pop_estimate_components_df = utils.load_population_projections_df()
```

```python
historical_components_df = utils.load_historical_estimates_df()
```

```python
bay_area_historical_components_df = historical_components_df.loc[lambda x: x['County'].isin(bay_area_counties)]
```

```python
bay_area_projections_df = pop_estimate_components_df.loc[lambda x: x['County'].isin(bay_area_counties)]
```

# Look at birth/death rates

```python
bay_area_historical_components_df[['County', 'Year']].assign(
    birth_rate=bay_area_historical_components_df['Births'] / bay_area_historical_components_df['Population (July 1)']
).set_index(['County', 'Year']).unstack().applymap('{:.2%}'.format)
```

```python
bay_area_historical_components_df[['County', 'Year']].assign(
    birth_rate=bay_area_historical_components_df['Deaths'] / bay_area_historical_components_df['Population (July 1)']
).set_index(['County', 'Year']).unstack().applymap('{:.2%}'.format)
```

```python
pop_estimate_components_df
```

```python
births_df = bay_area_projections_df[[
    'County', 'Start Year'
]].assign(
    birth_rate=bay_area_projections_df['Births'] / bay_area_projections_df['Start Population']
)

births_df.set_index(['County', 'Start Year']).unstack().applymap('{:.2%}'.format)
```

```python
%config InlineBackend.figure_format = 'retina'
alt.Chart(births_df, width=800, height=400, title='Birth rate predictions').mark_line().encode(
    x=alt.X('Start Year', title='Year', axis=alt.Axis(format='')),
    y=alt.Y('birth_rate', title='Birth rate', scale=alt.Scale(zero=False), axis=alt.Axis(format='%')),
    color='County'
)
```
